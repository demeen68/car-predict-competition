```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data_df = pd.read_csv('game_data/clean_data.csv')
```

```python
data_df.head()
```

# ONE HOT 编码CAR_MODEL （车型）

```python
data_df = pd.get_dummies(data_df,dummy_na=True,columns=['CAR_MODEL'])
```

```python
data_df.head()
```

# 正例数据生成(IS_LOST = 1,流失数据)

```python
data_df['IS_LOST'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
x = data_df.drop('IS_LOST',axis=1)
y = data_df['IS_LOST']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 0)
```

```python
oversampler = SMOTE(random_state=0)
ov_x,ov_y = oversampler.fit_sample(x_train,y_train)
```

```python
ov_x = pd.DataFrame(ov_x)
ov_y = pd.DataFrame(ov_y)
```

```python
data_df = ov_x.join(ov_y)
```

```python
data_df['IS_LOST'].value_counts()
```

```python
data_df.to_csv("game_data/preprocessing_data.csv",index=None)
```
