```python
len(df[df['CUST_AGE'].isnull()].index)
```

```python
df.describe()
```

### 全0填充 F_INSORNOT，ALL_BUYINS_N，DLRSI_CNT，GLASSBUYSEPARATE_CNT，SII_CNT

```python
df['F_INSORNOT'].fillna(0,inplace=True)
df['ALL_BUYINS_N'].fillna(0,inplace=True)
df['DLRSI_CNT'].fillna(0,inplace=True)
df['GLASSBUYSEPARATE_CNT'].fillna(0,inplace=True)
df['SII_CNT'].fillna(0,inplace=True)
```

```python
df.describe()
```

## 添加新维度 ONCE_PAY = CAR_PRICE - LOAN_AMOUNT

由于Loan_amount 与 car_price 方差太大，因此添加新维度表示买车时一次性支出的金额

```python
df['ONCE_PAY'] = df['CAR_PRICE'] - df['LOAN_AMOUNT']
df.head(3)
```

```python
sns.set(rc={"figure.figsize": (8, 4)})
x = df[df['ONCE_PAY']>0].ONCE_PAY
ax = sns.distplot(x)
```

```python
df.isnull().any()
```

### 偏度处理——开方

```python
df.describe()
```

```python
def check_skew(dataframe):
    skew_attention = []
    for column in dataframe.columns:
        if (dataframe[column].dtype == 'int64') or (dataframe[column].dtype == 'float64'):
            skew = stats.mstats.skew(dataframe[column]).data
            if skew >= 1:
                skew_attention.append(column)
    return skew_attention
```

```python
check_skew(df)
```

```python
df['CAR_AGE'] = np.sqrt(df['CAR_AGE'])
df['CAR_PRICE'] = np.sqrt(df['CAR_PRICE'])
df['LOAN_AMOUNT'] = np.sqrt(df['LOAN_AMOUNT'])
df['ONCE_PAY'] = np.sqrt(df['ONCE_PAY'])
```

ONCE_PAY 有一些值变为 Nan，全0填充

```python
df['ONCE_PAY'].fillna(0,inplace=True)
```

```python
df.describe()
```

```python
df.to_csv('game_data/clean_data.csv',index=None,)
```
